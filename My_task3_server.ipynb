```python
import socket
import numpy as np
import cv2
import threading

s = socket.socket() # creating socket with by-default settings(i.e.with IPv4 Address family and TCP protocol)  

s.setsockopt(socket.SOL_SOCKET,socket.SO_REUSEADDR,1) # To reuse the server

Port_No = 1234 # A random  number.
                     
                     #Binding the port no. with IP address of Local System.
s.bind(("",Port_No)) # Automatically detects the IP of local system and bind with port no.
s.listen() # Turning On listening mode. 
client_session, address = s.accept() # accepting request from the Server.
print(client_session.recv(4096)) # 4096 is buffer size
cameraIndex =0
camera = cv2.VideoCapture(cameraIndex) # Connecting local system's camera to This Code.

def sender():
    while True:
        status, photo = camera.read()             # camera starts capturing Video for Streaming
        photo = cv2.resize(photo, (640, 480))     #Reshaping the Frame size.
        print(photo.shape)
        if status:
            client_session.send(np.ndarray.tobytes(photo))    # converting Video's array into "byte" datatype to SEND Video
        else: print("Could not get frame")

def receiver():
    framesLost = 0
    print("Entered")
    while True:
        framesLost += 1
        data = client_session.recv(100000000)    # 10^8 buffer size for received video
        if(data == b'finished'):
            print("Finished")
            camera.release()                     # disconnecting the PC's Camera                   
            client_session.close()               # Cllosing the session with the client.
        else:
            photo =  np.frombuffer(data, dtype=np.uint8)      # Converting received "byte" data into "array" format 
            if len(photo) == 640*480*3:
                cv2.imshow('From Client->Web Cam ', photo.reshape(480, 640, 3))
                if cv2.waitKey(1) == 13:              # 13 for "Enter" key
                    client_session.send(b'finished')
                    camera.release()
                    cv2.destroyAllWindows()
                    break
            else:
                print("Lost {} frames".format(framesLost) )

threading.Thread(target=sender).start()
threading.Thread(target=receiver).start()
```
