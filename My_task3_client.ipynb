```python
import socket
import numpy as np
import cv2
import threading

s = socket.socket()
s.setsockopt(socket.SOL_SOCKET,socket.SO_REUSEADDR,1)
server_ip = input("Enter Server's IP addr: ")
server_port = input("Enter Server's Port No.: ")

s.connect((server_ip, int(server_port))) #  Requesting the server to connect.
s.send(b"connected")  # sending string as data
camera = cv2.VideoCapture("http://192.168.0.3:4747/video") # Using Mobile-App for Web-Cam.

# function for cleint to work as receiver
def receiver():
    framesLost = 0
    print("Entered")
    while True:
        framesLost += 1 # counting frame
        data = s.recv(100000000)  # receiving data with the size limit
        if(data == b'finished'): # to stop receiving and stop camera
            print("Finished")
            camera.release()
            s.close()
        else:  # converting the byte data into numpy array
            photo =  np.frombuffer(data, dtype=np.uint8)
            if len(photo) == 640*480*3: # changing the array shape and getting the video
                cv2.imshow('From Server->My Local System', photo.reshape(480, 640, 3))
                if cv2.waitKey(1) == 13: # camera closing condition (13==enter key)
                    s.send(b'finished')
                    camera.release()
                    cv2.destroyAllWindows()
                    break
            else:
                print("Lost {} frames".format(framesLost) ) #counting the lost frames during video exchange

# function for client to send the data of the video             
def sender():
    while True:  
        status, photo = camera.read()           # reading the camera data resizing and sending it as byte
        photo = cv2.resize(photo, (640, 480))    # for resizing frame, otherwise a long frame will come & keep on increasing. 
        print(photo.shape)
        if status:
            s.send(np.ndarray.tobytes(photo))  # converting Video's "array" into "byte" datatype
        else: print("Could not get frame")
    camera.release()

# Threading for sending and receiving simultaneously. 
threading.Thread(target=receiver).start() 
threading.Thread(target=sender).start()
```
